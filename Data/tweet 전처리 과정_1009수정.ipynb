```python
file_name = 'twitter_'
fileformat = '.txt'
filename = file_name + id + fileformat
```

```python
with open(filename, 'r', encoding = "utf-8") as f:
    tweets = pd.read_csv(f, sep = "\n", names = ['data'])
f.close()
```

```python
tweets['@#'] = tweets['lower'].replace(to_replace = "(@|#)", value = " ", regex = True)
```

```python
tweet_data = tweets['emoji']
```
